```python
import pandas as pd
import matplotlib.pyplot as plt


def plot_fidelity(
    csv_path: str,
    label: str | None = None,
    ax: plt.Axes | None = None,
    save_path: str | None = None,
) -> plt.Axes:
    """
    Load a fidelity CSV and plot F vs T/T_min.

    Parameters
    ----------
    csv_path  : Path to the input CSV (two columns: F, T_over_Tmin).
    label     : Legend label for this curve. Pass labels to both calls,
                then call ax.legend() afterward.
    ax        : Axes to plot on. Defaults to the current active axes
                (plt.gca()), so successive calls share the same plot.
    save_path : Optional path to save the figure (e.g. "plot.pdf").

    Returns
    -------
    ax : The Axes object, for further customisation.
    """
    plt.rcParams["text.usetex"] = False
    plt.rcParams.update({
        "font.family": "DejaVu Sans",
        "mathtext.fontset": "cm",
    })

    df = pd.read_csv(csv_path, header=None, names=["F", "T_over_Tmin"])
    df = (
        df.groupby("T_over_Tmin", as_index=False)["F"]
        .max()
        .sort_values("T_over_Tmin")
    )

    if ax is None:
        ax = plt.gca()

    ax.plot(df["T_over_Tmin"], df["F"], marker="o", label=label)
    ax.set_xlabel(r"$T/T_{min}$", fontsize=16)
    ax.set_ylabel(r"$F$", fontsize=16)
    ax.grid(True)

    if save_path:
        ax.get_figure().tight_layout()
        ax.get_figure().savefig(save_path)

    return ax
```

```python
plot_fidelity("./coupling_data/capacitiveCoup/Fidelities/qutrit_CZ_capacitiveCoup_M40_SGD_SRK2_g1.0_maxT2.0_maxD40_it5000.csv",label="capactive coup")
plot_fidelity("./coupling_data//capacitiveCoupUnit/Fidelities/qutrit_CZ_capacitiveCoupUnit_M40_SGD_SRK2_g1.0_maxT2.0_maxD40_it5000.csv",label="unit coup")
plot_fidelity("./coupling_data/capacitiveCoupExtra/Fidelities/qutrit_CZ_capacitiveCoupExtra_M40_SGD_CFME4_g1.0_maxT2.0_maxD40_it5000.csv",label="increased coup")
plot_fidelity("./coupling_data/capacitiveCoupMin/Fidelities/qutrit_CZ_capacitiveCoupMin_M40_SGD_SRK2_g1.0_maxT2.0_maxD40_it5000.csv",label="minimized coup")

ax = plt.gca()
ax.legend(fontsize=13)
plt.tight_layout()
plt.savefig("./Revision_Plots/CZ_Couplings.pdf")
plt.show()
```
