```python
drives = ["X1_01","X2_01","Y1_01","Y2_01","X1_12","Y1_12","X2_12","Y2_12"]
Weights = pd.read_csv("../../Paper2FinalData/capacitiveCoup/Weights/qutrit_CZ_capacitiveCoup_M40_SGD_SRK2_g1.0_maxT2.0_maxD40_it5000_Weights/Weights_t1.1.csv",header=None,names=drives)
Weights = Weights.append(Weights.iloc[-1],ignore_index=True)
```

```python
M = 20
ts = [m for m in range(M+1)]

plt.step(ts[:M+1],Weights[drives[0]][:M+1])
plt.xlabel('$m$',fontsize=20)
plt.ylabel('$\Omega^{x,1}_1/g$',fontsize=19)
plt.grid(which='major', linestyle='-', linewidth='0.5')
plt.grid(which='minor', linestyle='dotted', linewidth='0.5')
plt.minorticks_on()

plt.savefig("segments.pdf")
```
